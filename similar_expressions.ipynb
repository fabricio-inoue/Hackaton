```python
from gensim.models import Word2Vec
from itertools import combinations
import re

# Lista de expressões
expressions = ["inocência presumida", "falsa confissão", "conflito armado", "conflitos armados", "conflito civil", "conflitos civis", "guerra", "conflito", "democracia", "Convenção de Genebra", "tratado", "paz", "manutenção da paz", "corrupção", "instituição", "funcionário público", "governo", "suborno", "crime", "crimes", "criminoso", "déficit democrático", "democratização", "institucional", "tomada de decisão", "sociedade", "política", "ajuda financeira", "conflito étnico", "conflitos étnicos", "exoneração", "genocídio", "homicídio", "assassinato", "tráfico humano", "sistema de justiça criminal", "sistema de justiça", "justiça arbitrária", "refugiado", "terrorista", "violência", "tortura", "estado de direito efetivo", "fluxo de armas", "instituição transparente", "instituições transparentes", "boa governança", "identidade legal para todos", "liberdade de informação", "instituição de direitos humanos", "ativistas de direitos humanos", "liberdade fundamental", "liberdades fundamentais", "conflito violento", "conflitos violentos", "sociedade pacífica", "instituição efetiva", "instituições eficazes", "instituição responsável", "instituições responsáveis", "instituição inclusiva", "instituições inclusivas", "abuso infantil", "detenção arbitrária", "detenção sem sentença", "sistema judicial", "tribunal criminal", "sociedade inclusiva", "sociedades inclusivas", "instituição responsiva", "instituições responsivas", "sociedade justa", "sociedades justas", "recurso/remédio legal", "recursos/remédios legais", "independência do judiciário", "judiciário independente", "separação de poderes", "extremismo", "crime de guerra", "crime organizado", "transferência ilícita", "dinheiro ilícito", "tráfico de armas", "cibercrime / crime/s cibernético/s", "insurgência", "instituição democrática", "instabilidade política", "tomada de decisão política", "Convenção de Aarhus", "liberdade de imprensa", "liberdade de expressão"]

# Pré-processamento dos dados
expressions_processed = [re.sub(r'[^\w\s]', '', exp.lower()).split() for exp in expressions]

# Treinamento do modelo Word2Vec
model = Word2Vec(expressions_processed, vector_size=100, window=5, min_count=1, workers=4)

# Geração de expressões semelhantes
similar_expressions = []

for exp in expressions_processed:
    similar_words = set()
    for word in exp:
        similar_words.update([similar_word[0] for similar_word in model.wv.most_similar(word)])
    similar_expressions.append(similar_words)

# Impressão das expressões semelhantes
for i, exp in enumerate(similar_expressions):
    if i < 20:
        print(f"Expressão original: {expressions[i]}")
        print("Expressões semelhantes:")
        for similar_exp in combinations(exp, len(exp)):
            print(" ".join(similar_exp))
        print()
```
